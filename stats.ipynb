```python
stats = stats.drop('marking',axis=1)
stats.head()
```

```python
stats = stats.drop_duplicates()
```

```python
stats['value'].head()
```

```python
stats.loc[:, 'value'] = stats['value'].str.replace('$', '')
stats.loc[:, 'value'] = stats['value'].str.replace('.', '')
stats['value'].head()
```

```python
plt.hist(stats['height'], bins=30, color='purple', edgecolor='black')

# Adding title and labels
plt.title('Height Distribution')
plt.xlabel('Height')
plt.ylabel('Count')

# Display the plot
plt.show()
```

```python
# Plotting the histogram
plt.hist(stats['weight'], bins=30, color='purple', edgecolor='black')

# Adding title and labels
plt.title('Weight Distribution')
plt.xlabel('weight')
plt.ylabel('count')
```

```python
plt.hist(stats['age'], bins=30, color='purple', edgecolor='black')

# Adding title and labels
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```
